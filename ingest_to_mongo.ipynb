```python
"""
ingest_to_mongodb.py

Loads the cleaned Adult Income dataset into MongoDB.
"""

import pandas as pd
from pymongo import MongoClient
from pymongo.errors import ConnectionFailure

#MongoDB Configuration

MONGO_URI = "mongodb://admin:password@localhost:27017/"

DATABASE_NAME = "adult_income_db"rr
COLLECTION_NAME = "adult_income"

CSV_FILE = "data/adult-income-clean.csv"

#Connect to MongoDB

try:
    client = MongoClient(MONGO_URI)

    client.admin.command("ping")

    print("✓ Connected to MongoDB")

except ConnectionFailure:
    print("✗ Could not connect to MongoDB")
    raise

#Select Database

db = client[DATABASE_NAME]

collection = db[COLLECTION_NAME]

#Load Clean Dataset

print("Loading cleaned dataset...")

df = pd.read_csv(CSV_FILE)

print(f"Rows loaded: {len(df)}")

#Convert DataFrame

records = df.to_dict(orient="records")

#Clear Existing Data

deleted = collection.delete_many({})

print(
    f"Deleted {deleted.deleted_count} existing documents"
)

#Insert Documents

result = collection.insert_many(records)

print(
    f"Inserted {len(result.inserted_ids)} documents"
)

#Verification

count = collection.count_documents({})

print(f"Collection count: {count}")

#Sample Document

print("\nSample document:\n")

sample = collection.find_one({}, {"_id": 0})

print(sample)

#Close Connection

client.close()

print("\nIngestion complete.")
```
